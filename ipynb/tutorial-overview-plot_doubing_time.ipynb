### Doubling time

The sixth graph returned by the `overview` function 
(see the [overview](tutorial-overview-graphs.nbdata.html)
and the explanation of the [observables](tutorial-observables.nbdata.html))
shows the time over which the number of COVID cases and deaths doubles.
Here we show how it is created internally.

We call the function `plot_doubling_time` with the following parameters:
* `ax` — an axis object;
* `series` —  the data source;
* `color`  —  the color of the line, here we use the scheme `CN` where `N` is a number digit, but one may use [any colormap](https://matplotlib.org/tutorials/colors/colormaps.html);
* `labels`— a pair of strings: region label and identifier of cases or deaths.



```python
fig, ax = plt.subplots();    # create an empty graph
country = "Germany"
region = "Bremen"

cases, deaths = ov.get_country_data(country, region)    # get actual data

# pass the data to the plotting function
ov.plot_doubling_time(ax=ax, series=cases, labels=(f"{country}: {region}","cases"), color="C1");
# ov.plot_doubling_time(ax=ax, series=deaths, labels=(f"{country}: {region}","deaths"), color="C4");

ax.set_title("Doubling time:cases");
```

We also use some rather complicated tricks to scale the axes of the graph.

Since those formatting details are not essentias for the data analisys itself, let's
just show two graphs (cases and deaths) separately:

```python
fig, ax = plt.subplots();    # create an empty graph
country = "Germany"
region = "Bremen"

cases, deaths = ov.get_country_data(country, region)    # get actual data

# pass the data to the plotting function
# ov.plot_doubling_time(ax=ax, series=cases, labels=(f"{country}: {region}","cases"), color="C0");
ov.plot_doubling_time(ax=ax, series=deaths, labels=(f"{country}: {region}","deaths"), color="C4");

ax.set_title("Doubling time: deaths");
```
